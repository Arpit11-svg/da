```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Taken proxies for multiple IP addresses so the server not block us
proxies = {
  "https": "scraperapi.output_format=csv.autoparse=true.max_cost=98.keep_headers=true:ba23411677854bd5dddeb9a9e8f281f1@proxy-server.scraperapi.com:8001"
}
webpage= requests.get('https://www.ambitionbox.com/list-of-companies?page=3', proxies=proxies, verify=False)
print(webpage.text)
```

```python
soup=BeautifulSoup(webpage.text,'html.parser')
```

```python
print(soup.prettify()).text() #.prettify() =>format the content in correct way
```

```python
# to extract all tags from BeautifulSoup
h2_tags=soup.find_all('h2')
h2_tags
```

```python
for tag in h2_tags:
    title = tag.get_text(strip=True)  # strips whitespace
    print(title)
```

```python
ratings = soup.find_all('div', class_='rating_star_container')

for r in ratings:
    rating_text = r.text.strip()  # get the number like "3.6"
    print(rating_text)
```

```python
company=soup.find_all('div',class_='companyCardWrapper') # each company box comes individually
len(company)
```

```python
# TASK: fetch data of one page
name=[]
rating=[]
no_review=[]
sector=[]
locations=[]
for i in company:
  name.append(i.find('h2').text.strip())
  rating.append(i.find('div',class_='rating_star_container').text.strip()) # if we are using find_all().text=> gives error becuase in find_all it gives also link so how we can convert link into  text.
  no_review.append(i.find('span',class_='companyCardWrapper__companyRatingCount').text.strip().strip("()"))  # to eliminate extra braces
  sector.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[0].strip())
  locations.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[1].strip())
print(name)
print(rating)
print(no_review)
print(sector)
print(locations)

print(len(name))
print(len(rating))
print(len(no_review))
print(len(sector))
print(len(locations))
```

```python
dictionary={
    'name':name,
    'rating':rating,
    'no_review':no_review,
    'sector':sector,
    'locations':locations
}
df=pd.DataFrame(dictionary)
df
```

```python
import time
```

```python
# TASK: Now we have to do for multiple page
name=[]
rating=[]
no_review=[]
sector=[]
locations=[]
for j in range(1,60):
  url=f'https://www.ambitionbox.com/list-of-companies?page={j}'
  webpage= requests.get(url, proxies=proxies, verify=False)
  soup=BeautifulSoup(webpage.text,'html.parser')
  company=soup.find_all('div',class_='companyCardWrapper')
  for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('div',class_='rating_star_container').text.strip()) # if we are using find_all().text=> gives error becuase in find_all it gives also link so how we can convert link into  text.
    # no_review.append(i.find('span',class_='companyCardWrapper__companyRatingCount').text.strip().strip("()"))  # to eliminate extra braces
    reviews = i.find('span', class_='companyCardWrapper__companyRatingCount')
    no_review.append(reviews.text.strip().strip("()") if reviews else None)
    # sector.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[0].strip())
    # locations.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[1].strip())
    interlink = i.find('span', class_='companyCardWrapper__interLinking')
    if interlink:
        parts = [p.strip() for p in interlink.text.strip().split('|')]
        if len(parts) == 2:
            sector.append(parts[0])
            locations.append(parts[1])
        elif len(parts) == 1:
            sector.append(parts[0])
            locations.append(None)  # or ''
        else:
            sector.append(None)
            locations.append(None)
    else:
        sector.append(None)
        locations.append(None)
  time.sleep(1)
dictionary={
  'name':name,
  'rating':rating,
  'no_review':no_review,
  'sector':sector,
  'locations':locations
}
df=pd.DataFrame(dictionary)
df

```

```python
df_sorted = df.sort_values(by='name').reset_index(drop=True)
print(df_sorted)
```

```python
df_sorted.to_csv("ambitionbox_companies.csv", index=False)
```
