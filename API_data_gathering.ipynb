```python
import pandas as pd
import numpy as np
import requests
```

```python
api_key = input("Enter your TMDB API key: ").strip()
```

```python
url = f"https://api.themoviedb.org/3/movie/popular?api_key={api_key}&language=en-US&with_original_language=en&page=1"
```

```python
requests.get(url)
```

```python
response=requests.get(url).json()
```

```python
for i in response['results']:
  print(i['title'],",", "Vote Avg is " ,i['vote_average'])
```

```python
# Now, lets make DataFrame from this json
popularity=[]
vote_count=[]
original_language=[]
title=[]
vote_average=[]
overview=[]
release_date=[]

for i in response['results']:
  popularity.append(i['popularity'])
  vote_count.append(i['vote_count'])
  original_language.append(i['original_language'])
  title.append(i['title'])
  vote_average.append(i['vote_average'])
  overview.append(i['overview'])
  release_date.append(i['release_date'])

d={'title':title,
   'overview':overview,
   'original_language':original_language,
   'release_date':release_date,
   'popularity':popularity,
   'vote_count':vote_count,
   'vote_average':vote_average
   }

df=pd.DataFrame(d)
```

```python
df
```

```python
# Method 2 to form dataframe from json directly
df = pd.json_normalize(response['results'])
df = df[['title', 'overview', 'original_language', 'release_date', 'popularity', 'vote_count', 'vote_average']]
df
```

```python
import time
final=pd.DataFrame()
for j in range(1,350):
  url = f"https://api.themoviedb.org/3/movie/popular?api_key={api_key}&language=en-US&with_original_language=en&page={j}"
  response=requests.get(url).json()
  # print(f"Page {i} done")  # track progress
  time.sleep(0.25)  # 0.25 seconds delay to avoid hitting rate limit
  popularity=[]
  vote_count=[]
  original_language=[]
  title=[]
  vote_average=[]
  overview=[]
  release_date=[]

  for i in response['results']:
    popularity.append(i['popularity'])
    vote_count.append(i['vote_count'])
    original_language.append(i['original_language'])
    title.append(i['title'])
    vote_average.append(i['vote_average'])
    overview.append(i['overview'])
    release_date.append(i['release_date'])

  d={'title':title,
    'overview':overview,
    'original_language':original_language,
    'release_date':release_date,
    'popularity':popularity,
    'vote_count':vote_count,
    'vote_average':vote_average
    }
  df=pd.DataFrame(d)
  final=pd.concat([final,df],ignore_index=True)
```
