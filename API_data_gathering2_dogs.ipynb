```python
import pandas as pd
import requests
```

```python
api_key = "live_Fsb9H3kJ2LxZal25cHb8hWTjJkU2zWk74tAwOO1NoVrAhdZE8CAskjI0KuACyt4H"
url = f"https://api.thedogapi.com/v1/images/search?size=med&mime_types=jpg&format=json&has_breeds=true&order=RANDOM&page=0&limit=100&api_key={api_key}"
```

```python
for i in response:
    breed_info = i['breeds'][0]  # Access the first breed in the 'breeds' list
    print(breed_info['name'] + " and " + breed_info['life_span'] + " and " + i['url'])
```

```python
name=[]
life_span=[]
url=[]
bred_for=[]
breed_group=[]
height=[]
width=[]

for item in response:
    breed_info = item['breeds'][0]  # access the first breed in the breeds list
    name.append(breed_info['name'])
    life_span.append(breed_info['life_span'])
    url.append(item['url'])
    height.append(breed_info['height']['metric'])
    width.append(item['width'])
    bred_for.append(breed_info.get('bred_for',None))
    breed_group.append(breed_info.get('breed_group',None))
dict={
    'name':name,
    'life_span':life_span,
    'Photo\'s url':url,
    'bred_for':bred_for,
    'bred_group':breed_group,
    'height':height,
    'width':width
}

df=pd.DataFrame(dict)
```
